```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('ho-beo.jpg')
#plt.axis('off')
plt.imshow(img)
```

```python
img.shape
```

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
print(r.shape)
```

```python
grayscale = 0.3*r + 0.59*g + 0.11*b
```

```python
plt.imshow(grayscale, cmap='gray')
```

```python
m,n = grayscale.shape
output = np.zeros([m,n])
for x in range(m):
    for y in range(n):
        if grayscale[x,y]>127:
            output[x,y]=255
        else:
            output[x,y]=0
        
plt.imshow(output, cmap='gray')
```

```python
anh_gamma = plt.imread('fractured_spine.tif')
plt.subplot(221)
plt.imshow(anh_gamma,cmap='gray')
plt.title('mac dinh')
plt.axis('off')

plt.subplot(222)
plt.imshow(np.power(anh_gamma, 0.6),cmap='gray')
plt.title('gamma = 0.6')
plt.axis('off')


plt.subplot(223)
plt.imshow(np.power(anh_gamma, 0.2),cmap='gray')
plt.title('gamma = 0.2')
plt.axis('off')

plt.subplot(224)
plt.imshow(np.power(anh_gamma, 2.2),cmap='gray')
plt.title('gamma = 2.2')
plt.axis('off')
```
